```python
from langchain_core.documents import Document

# Assuming 'documents' is a list of existing Document objects
docs = []
for doc in documents:
    # Extract existing metadata
    existing_metadata = doc.metadata

    # Parse the page_content to extract 'First_Name' and 'Email'
    lines = doc.page_content.split('\n')
    metadata_updates = {}
    for line in lines:
        if line.startswith('First_Name:'):
            metadata_updates['First_Name'] = line.split(':', 1)[1].strip()
        elif line.startswith('Email:'):
            metadata_updates['Email'] = line.split(':', 1)[1].strip()

    # Update the metadata with the new information
    updated_metadata = {**existing_metadata, **metadata_updates}

    # Create a new Document object with updated metadata
    new_doc = Document(
        page_content=doc.page_content,
        metadata=updated_metadata
    )
    # Add the new Document to the docs list
    docs.append(new_doc)

# Example: Print the content and metadata of the first document in docs
print(docs[0].page_content)
print(docs[45].metadata)
```

```python
import pinecone
from sentence_transformers import SentenceTransformer
import numpy as np
import json
from langchain.schema import Document

# Load JSON data
with open("data.json", "r") as file:
    data = json.load(file)

# Define keys for page content and metadata
keys_for_content = [
    "Attendance (%)", "Midterm_Score", "Final_Score", "Assignments_Avg",
    "Quizzes_Avg", "Participation_Score", "Projects_Score",
    "Total_Score", "Grade", "Study_Hours_per_Week", "Parent_Education_Level"
]
keys_for_metadata = ["Gender", "Department"]

# Initialize Pinecone with API key
PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")
pc = pinecone.Pinecone(api_key=PINECONE_API_KEY)

# Load the sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create Document objects
documents = []
for record in data:
    # Create content string from available fields
    content_parts = []
    for key in keys_for_content:
        if key in record and record[key] is not None:
            content_parts.append(f"{key}: {record[key]}")
    content = ", ".join(content_parts)
    
    # Create metadata dictionary
    metadata = {
        'text': content,  # Store content in metadata for retrieval
        'gender': record.get('Gender', 'Unknown'),
        'department': record.get('Department', 'Unknown')
    }
    
    # Create Document object
    doc = Document(page_content=content, metadata=metadata)
    documents.append(doc)

# Set up Pinecone index
index_name = "meta-data"
index = pc.Index(index_name)

def store_data():
    """Stores data with metadata in Pinecone."""
    batch_size = 50
    for i in range(0, 200, batch_size):
        batch = documents[i:i+batch_size]
        vectors = []
        
        for doc in batch:
            # Create embedding
            embedding = model.encode(doc.page_content).astype(np.float32)
            # Create unique ID
            vector_id = str(abs(hash(doc.page_content)))[:16]
            # Add to batch with metadata (including the text)
            vectors.append((vector_id, embedding.tolist(), doc.metadata))
        
        # Upload batch
        index.upsert(vectors=vectors)
        print(f"Uploaded batch of {len(batch)} documents")
    
    return "Data stored successfully."

def filter_data(filter_metadata):
    """
    Filters data based on metadata.
    
    Args:
        filter_metadata (dict): Filter criteria (e.g., {"gender": "Male"})
    """
    dimension = model.get_sentence_embedding_dimension()
    results = index.query(
        vector=[0] * dimension,
        top_k=10,
        include_metadata=True,
        filter=filter_metadata
    )
    return results

if __name__ == "__main__":
    # First store the data


    # Then query for male students
    query_filter = {"gender": "Male"}
    filtered_results = filter_data(query_filter)
    
    # Print results
    print("\nQuery Results:")
    for match in filtered_results.matches:
        print(f"\nText: {match.metadata['text']}")
        print(f"Gender: {match.metadata['gender']}")
        print(f"Department: {match.metadata['department']}")
        print(f"Score: {match.score}")
```
